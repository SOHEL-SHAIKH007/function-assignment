7.What are the advantages of using generator over regular function

-> 1. Memory Efficiency:
Regular functions:
Store the entire result set in memory before returning, which can be problematic for large datasets.
Generators:
Produce values on demand using yield. They don't store the entire sequence, saving memory, especially when dealing with potentially huge or infinite sequences.
2. Lazy Evaluation:
Regular functions: Calculate and return the entire result at once.
Generators: Evaluate and produce values only when requested. This is called lazy evaluation or deferred execution, which can improve performance if not all values are needed.

```python
#4 Write python function that check if a given number is prime or not from 1 to 200
import math

def is_prime(num):

    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
# Example usage for numbers from 1 to 200
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number.")
    else:
        print(f"{number} is not a prime number.")
```

```python
#5 Create a iterator class in python that generates the fibonacci sequence up to a specified number of terms
class FibonacciIterator:
    def __init__(self, max_terms):
        if not isinstance(max_terms, int) or max_terms < 0:
            raise ValueError("max_terms must be a non-negative integer.")
        self._max_terms = max_terms
        self._a = 0
        self._b = 1
        self._count = 0

    def __iter__(self):

        return self

    def __next__(self):

        if self._count < self._max_terms:
            if self._count == 0:
                self._count += 1
                return 0
            elif self._count == 1:
                self._count += 1
                return 1
            else:
                next_fib = self._a + self._b
                self._a = self._b
                self._b = next_fib
                self._count += 1
                return next_fib
        else:
            raise StopIteration

```

```python
#6 Write a generator function that yeild the power of 2 up to a exponent
def powers_of_two(max_exponent):

    current_exponent = 0
    while current_exponent <= max_exponent:
        yield 2 ** current_exponent
        current_exponent += 1
```

```python
#7 Implement a gnerator function that reads a file line by line and yeild each line as a string
def read_file_line_by_line(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example Usage:
if __name__ == "__main__":
    with open("sample.txt", "w") as f:
        f.write("This is line 1.\n")
        f.write("This is line 2.\n")
        f.write("And this is line 3.")
    print("Reading lines from 'sample.txt':")
    for line in read_file_line_by_line("sample.txt"):
        print(line)

    print("\nAttempting to read a non-existent file:")
    for line in read_file_line_by_line("non_existent_file.txt"):
        print(line)
```

```python
#9 Write a python  program that uses 'map()' to convert a list of temperature celcius to fehrenheit
def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32
celsius_temperatures = [0, 10, 25, 37, 100]
fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)

fahrenheit_list = list(fahrenheit_temperatures)

print(f"Celsius temperatures: {celsius_temperatures}")
print(f"Fahrenheit temperatures: {fahrenheit_list}")
```

```python
#10 Create a python program uses 'filter()' to remove all the vowels from the string
def remove_vowels_filter(input_string):
  vowels = "aeiouAEIOU"
  filtered_characters = filter(lambda char: char not in vowels, input_string)
  result_string = "".join(list(filtered_characters))
  return result_string

# Example usage
my_string = "Hello World! This is a test."
string_without_vowels = remove_vowels_filter(my_string)

print(f"Original string: {my_string}")
print(f"String without vowels: {string_without_vowels}")
```
